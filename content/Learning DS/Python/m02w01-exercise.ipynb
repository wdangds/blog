## I. Linear algebra operations with NumPy

#### 1. Length of a vector

- Vector: **v** = $[v_1, v_2,...,v_n]^T$
- Length of a vector: $\mid\mid \text{\textbf{v}} \mid \mid =\sqrt{v_1^2+v_2^2+...+v_n^2}$

```python
import numpy as np

def compute_vector_length(vector):
    # ********** Your code here **********
    len_of_vector = np.linalg.norm(vector)
    return len_of_vector

vector = np.array([-2, 4, 9, 21])
result = compute_vector_length(vector)
print(round(result, 2))
```

#### 2. Dot product

- Vector:
$\displaystyle
    \mathbf{v} =
    \begin{bmatrix}
      v_1 \\
      v_2 \\
      \vdots \\
      v_n
    \end{bmatrix}
    \quad,\quad
    \mathbf{u} =
    \begin{bmatrix}
      u_1 \\
      u_2 \\
      \vdots \\
      u_n
    \end{bmatrix}
  $

- Dot Product:
  $\displaystyle
    \mathbf{v}\cdot\mathbf{u}
    = v_1\,u_1 \;+\; v_2\,u_2 \;+\;\dots+\;v_n\,u_n
  $


```python
def compute_dot_product(vector1, vector2):
    # ******** Your code here ********
    result = vector1.dot(vector2)
    # result = np.dot(vector1, vector2)
    return result

v1 = np.array([0, 1, -1, 2])
v2 = np.array([2, 5, 1, 0])
result = compute_dot_product(v1, v2)
print(round(result, 2))
```

#### 3. Multiplying a vector by a matrix

- Matrix: 
$\displaystyle
    \mathbf{A} = 
    \begin{bmatrix}
    a_{11} & \cdots & a_{1n}\\
    \cdots & \cdots & \cdots \\
    a_{m1} & \cdots & a_{mn}
    \end{bmatrix}, \mathbf{A} \in \mathbb{R}^{m\times n}
$

- Vector:
$\displaystyle
    \mathbf{v} = 
    \begin{bmatrix}
    v_1\\
    v_2\\
    \cdots\\
    v_n
    \end{bmatrix}, \mathbf{v} \in \mathbb{R}^n
$

- $\displaystyle
    \mathbf{c} =\mathbf{Av} =
    \begin{bmatrix}
    a_{11}\times v_1 + \cdots + a_{1n}\times v_{n}\\
    \cdots\\
    a_{m1}\times v_1 + \cdots + a_{mn} * v_{n}
    \end{bmatrix}, \mathbf{c}\in \mathbb{R}^m
$



```python
def matrix_multi_vector(matrix, vector):
    # ***** Your code here *****
    result = matrix @ vector
    return result

matrix = np.array([[-1, 1, 1],
                   [0, -4, 9]])
vector = np.array([0, 2, 1])
result = matrix_multi_vector(matrix, vector)
print(result)
```

#### 4. Multiplying a matrix by a matrix

- Matrix A:
$\displaystyle
    \mathbf{A} = 
    \begin{bmatrix}
        a_{11} & \cdots & a_{1n}\\
        \cdots & \cdots & \cdots \\
        a_{m1} & \cdots & a_{mn}
    \end{bmatrix}, \mathbf{A} \in \mathbb{R}^{m\times n}
$

- Matrix B:
$\displaystyle
    \mathbf{B} = 
    \begin{bmatrix}
        b_{11} & \cdots & b_{1k}\\
        \cdots & \cdots & \cdots \\
        a_{n1} & \cdots & a_{nk}
    \end{bmatrix}, \mathbf{A} \in \mathbb{R}^{n\times k}
$

- 
$\displaystyle
    \mathbf{C} = \mathbf{AB} =
    \begin{bmatrix}
        a_{11}\times b_{11} + \cdots + a_{1n}\times b_{n1} &\cdots&a_{11}\times b_{1k}+a_{1n} \times b_{nk}\\
        \cdots & \cdots &\cdots\\
        a_{m1}\times b_{11} + \cdots + a_{mn}\times b_{n1} & \cdots & a_{m1}\times b_{1k} + a_{mn}\times b_{nk}
    \end{bmatrix}, \mathbf{C} \in \mathbb{R}^{m\times k}
$


```python
import numpy as np
def matrix_multi_matrix(matrix1, matrix2):
    # ***** Your code here *****
    result = matrix1 @ matrix2
    return result

m1 = np.array([[0, 1, 2],
               [2, -3, 1]])
m2 = np.array([[1, -3],
               [6, 1],
               [0, -1]])
result = matrix_multi_matrix(m1, m2)
print(result)
```

## II. Cosine Similarity

- Data (vector x, y): $x=\{x_1, \cdots, x_N\}, y = \{y_1, \cdots, y_N\}$
- Cosine Similarity: $cs(\mathbf{x},\mathbf{y})=\frac{\mathbf{x}\cdot \mathbf{y}}{\mid\mid \mathbf{x}\mid \mid \mid \mid \mathbf{y}\mid \mid}=\frac{\sum_i^n x_i y_i}{\sqrt{\sum_i^n x_i^2}\sqrt{\sum_i^n y_i^2}}$

*Example*: Given that 
$\mathbf{x}=
\displaystyle
\begin{bmatrix}
1\\2\\3\\4
\end{bmatrix}, \mathbf{y}=
\begin{bmatrix}
1\\0\\3\\0
\end{bmatrix}
$
. Find the Cosine Similarity cs(x, y)

$\mid \mid \mathbf{x}\mid \mid = \sqrt{1^2+2^2+3^2+4^2}=\sqrt{30}, \mid \mid \mathbf{y} \mid \mid = \sqrt{1^2+0^2+3^2+0^2}= \sqrt{10}$

$\mathbf{x}\cdot\mathbf{y}= 1\times 1 + 2\times 0 + 3\times 3+ 4\times 0= 10$

$\Rightarrow cs(\mathbf{x},\mathbf{y})=\frac{10}{\sqrt{30}\times \sqrt{10}}=\frac{\sqrt{3}}{3}$

```python
def compute_cosine(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    return dot_product/ (norm_v1*norm_v2)

x = np.array([1, 2, 3, 4])
y = np.array([1, 0, 3, 0])
result = compute_cosine(x, y)
print(round(result, 3))
```

## IV. Image Processing 

### 1. Background Subtraction

Write a program using background subtraction technique to extract foreground (desired object) and paste into a new background. I/O of the program is:
- Input: 3 pictures - Orignal Background Image (Greenscreen), Target Background Image, and Object Image
- Ouput: New image after extract object from Object Image and paste into Target Background Image.

**Step 1**: Resize image to the same size

```python
# ! pip install opencv-python
import cv2
import matplotlib.pyplot as plt

def read_and_resize(raw_image):
    image = cv2.imread(raw_image, 1) # 1 for color, 0 for grayscale
    image = cv2.resize(image, (678, 381))
    return image

bg1_image = read_and_resize("GreenBackground.png")
ob_image = read_and_resize("Object.png")
bg2_image = read_and_resize("NewBackground.jpg")

ob_rgb = cv2.cvtColor(ob_image, cv2.COLOR_BGR2RGB)
plt.imshow(ob_rgb)
plt.show()
```

**Step 2**: Calculate the difference between the Object Image and the Original Background Image to obtain an intial mask.

```python
def compute_difference(bg_img, input_img):
    # ***** Your code here *****
    difference_single_channel = input_img - bg_img
    return difference_single_channel

difference_single_channel = compute_difference(bg1_image, ob_image)
print(difference_single_channel.shape)
dsc_rgb = cv2.cvtColor(difference_single_channel, cv2.COLOR_BGR2RGB)
plt.imshow(dsc_rgb)
plt.show()
```

**Step 3**: Compute Binary Mask - Convert the difference image into a binary mask (Foreground Mask). Pixels corresponding to the object are typically 1 (or 255), and background pixels are 0.

```python
def compute_binary_mask(difference_single_channel):
    # ***** Your code here *****
    threshold = 0
    binary_mask = (difference_single_channel>threshold).astype(np.uint8) * 255
    return binary_mask

binary_mask = compute_binary_mask(difference_single_channel)
print(binary_mask.dtype)
bm_rgb = cv2.cvtColor(binary_mask, cv2.COLOR_BGR2RGB)
plt.imshow(bm_rgb)
plt.show()
```

**Step 4**: Create a final output image by taking pixel values from the Object Image where the mask is 1 (or 255) and from the Target Background Image where the mask is 0.

```python
def replace_background(bg1_image, bg2_image, ob_image):
    difference_single_channel = compute_difference(
        bg1_image,
        ob_image
    )

    binary_mask = compute_binary_mask(difference_single_channel)

    output = np.where(binary_mask==255, ob_image, bg2_image)
    return output
```

```python
output = replace_background(bg1_image, bg2_image, ob_image)

output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
plt.imshow(output_rgb)
plt.show()
```

### 2. Converting Color Image to Grayscale

Converts a 3-channel image to a 1-channel grayscale image by combining RGB values into a single value. Three common methods are:
- **Lightness**: $\frac{max(R,G,B)+min(R,G,B)}{2}$
- **Average**: $\frac{R+G+B}{3}$
- **Luminosity**: $0.21*R+0.72*G+0.07*B$

```python
import matplotlib.image as mpimg

img = mpimg.imread("dog.jpeg")
plt.imshow(img)
plt.show()
```

1. Lightness technique

```python
# Using Lightness = (max(R,G,B) + min(R,G,B))/2
def lightness(img):
    return((np.max(img)+np.min(img))/2)
gray_img_01 = np.apply_along_axis(lightness, 2, img)
# gray_img_01 = (img.max(axis = 2) + img.min(axis = 2))/2
print(gray_img_01[0,0])

plt.imshow(gray_img_01)
plt.show()
```

2. Average technique

```python
# Using average: (R+G+B)/3
gray_img_02 = np.average(img, axis = 2)
print(gray_img_02[0, 0])

plt.imshow(gray_img_02)
plt.show()
```

3. Luminosity technique

```python
# Using luminosity: 0.21*R + 0.72*G + 0.07*B
def luminosity(chanel):
    return chanel[0]*0.21 + chanel[1]*0.72 + chanel[2]*0.07

gray_img_03 = np.apply_along_axis(luminosity, 2, img)
print(gray_img_03[0, 0])

plt.imshow(gray_img_03)
plt.show()
```
