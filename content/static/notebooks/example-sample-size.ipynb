## Example: Average number of Duke games attended

Let's look at the population data for the number of basketball games attended:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(12)
# This is NOT a lognormal distribution
data = np.random.lognormal(mean=2.0, sigma=1.0, size=250).astype(int)
data = np.clip(data, 0, 70)
```

```python
fig, ax = plt.subplots(figsize = (6,4))
ax.hist(data, bins = 7, edgecolor = 'black', alpha = 0.5)
ax.set_xlabel("Number of Duke games attended")
ax.set_ylabel("Frequency")

# annotate mu and sigma
txt = rf"$\mu = {np.mean(data):.2f}$" + "      " + rf"$\sigma = {np.std(data):.2f}$"
fig.text(0.5, -0.05, txt, ha="center", va="top", fontsize=14)

plt.show()
```

#### Effect of increasing sample size

```python
num_sample = [10, 30, 70]
fig, axes = plt.subplots(3, figsize=(10,12))
for ax, num in enumerate(num_sample):
    mu = []
    sigma = []
    for _ in range(100):
        sample = np.random.choice(data, size = num)
        mu.append(np.mean(sample))
        sigma.append(np.std(sample, ddof=0))
    
    axes[ax].hist(mu, edgecolor = 'black', alpha = 0.5)
    axes[ax].set_xlabel(rf"Sample size: {num}, $\mu = {np.mean(mu):.2f}$" + "      " + rf"$\sigma = {np.mean(sigma):.2f}$")
```
