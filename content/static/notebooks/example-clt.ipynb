## Example: coin flips

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters
n_trials = 10000 # number of experiments
n = 30           # sample size per experiment
```

```python
# coin flips (Bernoulli, p = 0.5)
p = 0.5
samples = np.random.binomial(1, p, size = (n_trials, n))
```

```python
# compute sums and averages
sums = samples.sum(axis = 1)
averages = samples.mean(axis = 1)
```

```python
# plot histograms
fig, ax = plt.subplots(1, 2, figsize=(12,5))
ax.flatten()

# sum distribution
ax[0].hist(sums, edgecolor = 'black', density = True, alpha = 0.7, color = 'blue')
ax[0].set_title(rf"Distribution of Sum $S_n$ (n={n})")
ax[0].set_xlabel("Sum of coin flips")
ax[0].set_ylabel("Density")

# average distribution
ax[1].hist(averages,edgecolor = 'black', density= True, alpha = 0.7, color = 'green')
ax[1].set_title(rf"Distribution of Average $\bar{{X}}_n$ (n={n})")
ax[1].set_xlabel("Average of coin flips")
ax[1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```
